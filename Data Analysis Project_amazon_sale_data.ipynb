### Exploratory Data Analysis

```python
#ploting the between Size of clothes and count of size-item
sns.countplot(x='Size',data=df)
```

```python
#ploting the between Size of clothes and count of size-item (using the actual value of count of size)
ax=sns.countplot(x='Size',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

Note : From above graph you can see that most of the people buys M-size

```python
#grouping the column 'Size' and 'Qty' and arrange in descending order of total quantity
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
#ploting the graph between 'Size' and total 'Quality'
s_qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=s_qty)
```

#### Courier status

```python
#ploting the graph 'Courier Status' and 'count of Courier Status'
sns.countplot(data=df,x='Courier Status',hue='Status')
```

```python
#histogram of 'Size' column of dataframe
plt.hist(df.Size)
plt.grid()
```

```python
#plot the histogram of 'category' column of dataframe 
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))  #canvas size
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

Note: From above graph you can see that most of the buyers are t-shirt

```python
#checking B2B Date by using pie chart
B2B_check=df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_check,labels=B2B_check.index,autopct="%1.1f%%")
plt.legend()
#plt.axis('equal')
plt.show()
```

Note: From Above chart we Can see that maximum 99.2% of buyers are retailers and 0.8% are B2B buyers

```python
#prepare data for scatter plot
x_data=df['Category']
y_data=df['Size']

#plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title("Scatter")
plt.title("Scatter Plot")
plt.show()
```

```python
#plot count of cities by all states
plt.figure(figsize=(20,10))
sns.countplot(data=df,x='ship-state')
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title('Distribution of  State')
plt.xticks(rotation=90)
plt.show()
```

```python
#plot count of cities by Top 10 states
top_10_state=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(20,10))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title('Distribution of  State')
plt.xticks(rotation=90)
plt.show()
```

Note: From above graph you can see that most of the buyers are maharastra state
