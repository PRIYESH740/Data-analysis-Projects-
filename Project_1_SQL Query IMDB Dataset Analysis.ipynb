# 1.Establishing the connection to the data base

```python
db="movies.sqlite"
# Connect to SQLite database
connect = sqlite3.connect(db)
curs=connect.cursor()
```

a cursor object is used to interact with the database. The cursor acts as a mediator between the SQLite database connection and your Python code. It allows you to execute SQL queries, fetch data, and manage transactions.

1. Name all the directors with the numbers of movies and revenue where revenue should be taken into account forr doing the analysis. The director who has the highest revenue should comes at the top and so and so for

```python
Top_directors=pd.DataFrame(curs.execute("SELECT directors.name,COUNT(*) as 'No. Of Movies',SUM(revenue) FROM movies JOIN directors ON movies.director_id=directors.id GROUP BY name ORDER BY sum(revenue) DESC"),columns=['Director','Total Movies',"Revenue"
])
```

2.Name all the director with the number of movies and revenue where number of movies should be taken into account for doing the analysis.The director who has the number of movies should comes at the top and so and so for

```python
Top_directors=pd.DataFrame(curs.execute("SELECT directors.name,COUNT(*) as 'No. Of Movies',SUM(revenue) FROM movies JOIN directors ON movies.director_id=directors.id GROUP BY name ORDER BY COUNT(*) DESC"),columns=['Director','Total Movies',"Revenue"
])
```
